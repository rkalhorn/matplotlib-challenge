```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load) 

# Combine the data into a single dataset
drugData_df=pd.merge(mouse_df,clinical_df, on="Mouse ID")

# Display the data table for preview
drugData_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
meanTumorVol = drugData_df.groupby(["Drug","Timepoint"], as_index=False)["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
meanTumorVol_df = pd.DataFrame(data = meanTumorVol)
meanTumorVol_df = meanTumorVol_df.rename(columns={"Tumor Volume (mm3)": "Mean Tumor Volume"})


# Preview DataFrame
meanTumorVol_df[0:12]
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
#Can't use as_index=False here becuase there is a bug in python https://github.com/pandas-dev/pandas/issues/16799
#instead use reset_index() after
semTumorVol = drugData_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem().reset_index()

# Convert to DataFrame
semTumorVol_df = pd.DataFrame(data = semTumorVol)
semTumorVol_df = semTumorVol_df.rename(columns={"Tumor Volume (mm3)": "SEM for Tumor Volume"})


# Preview DataFrame
semTumorVol_df.head()
```

```python
# Minor Data Munging to Re-Format the Mean Tumor Vol Data Frame
meanTumorVolData_reformatted = meanTumorVol_df.pivot(index="Timepoint",columns="Drug",values="Mean Tumor Volume")

# Preview that Reformatting worked
meanTumorVolData_reformatted

```

```python
# Minor Data Munging to Re-Format the SEM Tumor Vol Data Frame
semTumorVolData_reformatted = semTumorVol_df.pivot(index="Timepoint",columns="Drug",values="SEM for Tumor Volume")

# Preview that Reformatting worked
semTumorVolData_reformatted.head()
```

```python
# Generate the Plot (with Error Bars)
times = meanTumorVolData_reformatted.index

drugs = ["Capomulin","Infubinol", "Ketapril", "Placebo"]
marker = {"Capomulin":["o","r"],"Infubinol":["s","b"], "Ketapril":["^","g"], "Placebo":["d","k"]}

#format and label plot
plt.figure(figsize = (10,8))
plt.title("Mean Tumor Volume Response to Treatment Over Time")
plt.xlabel("Time (Days)")
plt.ylabel("Mean Tumor Volume (mm3)")
plt.grid()


for drug in drugs:
   plt.errorbar(times, meanTumorVolData_reformatted[drug],semTumorVolData_reformatted[drug], 
                linestyle = "--", color = marker[drug][1],marker=marker[drug][0])


plt.legend()
plt.show()
print(times)
```

```python
# Show the Figure
plt.show()
```

Observable Trend Concerning Average Tumor Volume:
The mice that were given Capomulin had a decrease in the avereage tumor volume over the 45 days. Whereas, the mice that were given Infubinol, Ketapril, and the Placebo had an increase in the average tumor volume over the 45 days. This may suggest that Capomulin decreases tumor volume whereas the other two drugs, Infubinol, Ketapril had little to no effect on average tumor volume since they both behaved very similiar to the placebo. In fact, since the error bars overlap the Placebo error bars for both the drugs, Inubinol and Ketapril, there is not a statistically significantly different result from the two drugs and the placebo over the 45 days. 

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
meanMetSiteData = drugData_df.groupby(["Drug","Timepoint"], as_index=False)["Metastatic Sites"].mean()

# Convert to DataFrame
meanMetSiteData_df = pd.DataFrame(data = meanMetSiteData)
meanMetSiteData_df = meanMetSiteData_df.rename(columns={"Metastatic Sites": "Mean Metastatic Sites"})

# Preview DataFrame
meanMetSiteData_df[0:12]

```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
semMetSiteData = drugData_df.groupby(["Drug","Timepoint"])["Metastatic Sites"].sem().reset_index()


# Convert to DataFrame
semMetSiteData_df = pd.DataFrame(data = semMetSiteData)
semMetSiteData_df = semMetSiteData_df.rename(columns={"Metastatic Sites": "SEM for Metastatic Sites"})

# Preview DataFrame
semMetSiteData_df.head()


```

```python
# Minor Data Munging to Re-Format the Mean Metastatic Site Data Frame
meanMetSiteData_reformatted = meanMetSiteData_df.pivot(index="Timepoint",columns="Drug",values="Mean Metastatic Sites")

# Preview that Reformatting worked
meanMetSiteData_reformatted
```

```python
# Minor Data Munging to Re-Format the SEM Metastatic Site Data Frame
semMetSiteData_reformatted = semMetSiteData_df.pivot(index="Timepoint",columns="Drug",values="SEM for Metastatic Sites")

# Preview that Reformatting worked
semMetSiteData_reformatted.head()
```

```python
# Generate the Plot (with Error Bars)

#format and label plot
plt.figure(figsize = (10,8))
plt.title("Mean Metastatic Sites over Time")
plt.xlabel("Time (Days)")
plt.ylabel("Mean Metastatic Sites")
plt.grid()


for drug in drugs:
   plt.errorbar(times, meanMetSiteData_reformatted[drug],semMetSiteData_reformatted[drug], 
                linestyle = "--", color = marker[drug][1],marker=marker[drug][0])


plt.legend()
plt.show()
```

Observable Trend Concerning the Average of Metastatic Sites:
While all treatments saw an increase in average number of metastatic sites over the 45 days, with the Placebo having the highest average metastatic sites at all measured times. But, by day 45 only Capomulin and Infubinol had statistically significant lower average metastatic sites than the placebo. This would suggest that both Capomulin and Infubinol may reduce the average number of metastatic sites than being unmedicated, while Ketapril had statistically no effect. 


## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouseCount = drugData_df.groupby(["Drug","Timepoint"], as_index=False)["Mouse ID"].count()

# Convert to DataFrame
mouseCount_df = pd.DataFrame(data = mouseCount)
mouseCount_df = mouseCount_df.rename(columns={"Mouse ID": "Mouse Count"})

# Preview DataFrame
mouseCount_df[0:12]
```

```python
# Minor Data Munging to Re-Format the Mouse Count Data Frame
mouseCountData_reformatted = mouseCount_df.pivot(index="Timepoint",columns="Drug",values="Mouse Count")

# Preview that Reformatting worked
mouseCountData_reformatted
```

```python
#format and label plot
plt.figure(figsize = (10,8))
plt.title("Mouse Count over Time")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.grid()


for drug in drugs:
   plt.plot(times, (mouseCountData_reformatted[drug]/mouseCountData_reformatted[drug][0]*100), 
                linestyle = "--", color = marker[drug][1],marker=marker[drug][0])


plt.legend(loc=3)
plt.show()
```

Observable Trend Concerning the Survival Rate Over 45 days:
While all treatments saw the survival percentage decrease over the 45 days, Ketapril and Infubinol ended with a survival rate equal to or less than the placebo. Therefore, this data does not provide evidence that Ketapril or Infubinol increase the survival rate over a 45 day period. Capomulin, on the other hand, has a significantly higher survival rate than the placebo, suggesting that it does decrease mortality over a 45 day period.

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
beginningVol = meanTumorVolData_reformatted.iloc[0]
endingVol = meanTumorVolData_reformatted.iloc[-1]
percentChange = ((endingVol-beginningVol)/beginningVol)*100
percentChange
```

```python
# Splice the data between passing and failing drugs
passingDrugs = percentChange[percentChange>0].loc[drugs].dropna(how="any")
failingDrugs  = percentChange[percentChange<=0].loc[drugs].dropna(how="any")



x_axis = np.arange(len(drugs))

# Orient widths. Add labels, tick marks, etc. 
plt.xlabel("Treatment")
plt.ylabel("% Mean Tumor Volume Change")
plt.title("Tumor Change Over 45 Day Treatment")
plt.xticks(x_axis, drugs)
plt.grid(axis="y")

# create plot

rects2 = plt.bar(np.arange(len(failingDrugs)),  list(failingDrugs.values),width = 1, edgecolor="k", align = "center",color = "g")
rects1 = plt.bar(len(failingDrugs)+np.arange(len(passingDrugs)), list(passingDrugs.values), width = 1,edgecolor="k",align = "center",color = "r")

for drug in drugs:
    if drug in list(failingDrugs.index):
        plt.text(list(failingDrugs.index).index(drug),-7, "{:,.2f}%".format(failingDrugs[drug]),ha="center", color = "w")
    elif drug in list(passingDrugs.index):
        plt.text(len(failingDrugs)+list(passingDrugs.index).index(drug),4, "{:,.2f}%".format(passingDrugs[drug]), ha="center", color = "w")
    
plt.show()
```
